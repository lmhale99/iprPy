```python
import iprPy
import numpy as np
from MeltCommander import MeltCommander

import pandas as pd
```

```python
database = iprPy.load_database('master')
lammps_command = '/home/lmh1/LAMMPS/2022-06-23/src/lmp_serial'
```

```python
melts_df = database.get_records_df('calculation_melting_temperature')

df = []

for family in np.unique(melts_df.family):
    fam_df = melts_df[melts_df.family == family]
    
    for potential_LAMMPS_id in np.unique(fam_df.potential_LAMMPS_id):
        pot_df = fam_df[fam_df.potential_LAMMPS_id == potential_LAMMPS_id]

        for composition in np.unique(pot_df.composition):
            comp_df = pot_df[pot_df.composition == composition]

            commander = MeltCommander(database, lammps_command, potential_LAMMPS_id, composition, family, verbose=False)

            data = {}
            data['family'] = family
            data['potential_LAMMPS_id'] = potential_LAMMPS_id
            data['composition'] = composition
            data['num_melts'] = len(commander.temperature_melts)
            if data['num_melts'] > 0:
                data['T_melt_mean'] = np.mean(commander.temperature_melts)
                data['T_melt_std'] = np.std(commander.temperature_melts) / (data['num_melts']**0.5)
            df.append(data)

df = pd.DataFrame(df)
```
