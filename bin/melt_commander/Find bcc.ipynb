# Find bcc

```python
import iprPy
import atomman as am
```

```python
database = iprPy.load_database('master')
```

```python
bcc_crystals, bcc_crystals_df = database.get_records('relaxed_crystal', family='A2--W--bcc', method='dynamic', standing='good', return_df=True)
len(bcc_crystals)
```

```python
eam_potentials_df = database.get_records_df('potential_LAMMPS', pair_style=['eam', 'eam/alloy', 'eam/fs'])
```

```python
bcc_eam_df = bcc_crystals_df[bcc_crystals_df.potential_LAMMPS_id.isin(eam_potentials_df.id)]
```

```python
bcc_eam_df.keys()
```

```python
keys = ['potential_LAMMPS_id', 'composition', 'a']
bcc_eam_df[keys].sort_values(keys).to_csv('Melting_temperatures_bcc.csv', index=False)
```
